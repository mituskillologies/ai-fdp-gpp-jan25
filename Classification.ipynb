```python
import pandas as pd
```

```python
data = pd.read_csv('ingredients.csv')
data
```

```python
# input data
x  = data[['sweetness','crunchiness']]

# output data
y = data['class']
```

```python
# import the class
from sklearn.tree import DecisionTreeClassifier

# create the object
tree = DecisionTreeClassifier(criterion='entropy')

# train the algorithm
tree.fit(x, y)
```

```python
tomato = pd.DataFrame({'sweetness':[9],'crunchiness':[6]})
```

```python
tomato
```

```python
tree.predict(tomato)
```

```python
from sklearn.tree import plot_tree
plot_tree(tree, fontsize=6, feature_names=x.columns);
```
